### Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

### Data Exploration and Visualization

```python
df = pd.read_csv("PJME_hourly.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df["Datetime"] = pd.to_datetime(df["Datetime"])
```

```python
df.set_index("Datetime", inplace=True)
df.head()
```

```python
plt.figure(figsize=(15, 5))
plt.plot(df.index, df['PJME_MW'], color='blue', alpha=0.7)
plt.title("PJME Electricity Load")
plt.xlabel("Datetime")
plt.ylabel("Electricity Load (MW)")
plt.grid()
plt.show()
```

```python
# Resampling the data for different time frames
df_yearly = df["PJME_MW"].resample('Y').mean()
df_monthly = df["PJME_MW"].resample('M').mean()
df_daily = df["PJME_MW"].resample('D').mean()

plt.figure(figsize=(18, 12))

# Yearly trends
plt.subplot(3, 1, 1)
plt.plot(df_yearly.index, df_yearly, label="Yearly Average Load", color='green')
plt.title("Yearly Electricity Load Trends")
plt.xlabel("Year")
plt.ylabel("Average Load (MW)")
plt.legend()
plt.grid()

# Monthly trends
plt.subplot(3, 1, 2)
plt.plot(df_monthly.index, df_monthly, label="Monthly Average Load", color='orange')
plt.title("Monthly Electricity Load Trends")
plt.xlabel("Month")
plt.ylabel("Average Load (MW)")
plt.legend()
plt.grid()

# Daily trends
plt.subplot(3, 1, 3)
plt.plot(df_daily.index, df_daily, label="Daily Average Load", color='purple')
plt.title("Daily Electricity Load Trends")
plt.xlabel("Day")
plt.ylabel("Average Load (MW)")
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

```python
# Extracting hour, day of the week, and month from the index for grouping
df['Hour'] = df.index.hour
df['DayOfWeek'] = df.index.day_name()  
df['Month'] = df.index.month_name()    

# Mapping full month names to their abbreviations
month_shortcuts = {
    'January': 'Jan', 'February': 'Feb', 'March': 'Mar', 'April': 'Apr',
    'May': 'May', 'June': 'Jun', 'July': 'Jul', 'August': 'Aug',
    'September': 'Sep', 'October': 'Oct', 'November': 'Nov', 'December': 'Dec'
}
df['Month'] = df['Month'].map(month_shortcuts)

# Calculating average consumption by hour
hourly_avg = df.groupby('Hour')['PJME_MW'].mean()

# Calculating average consumption by day of the week
dayofweek_avg = df.groupby('DayOfWeek')['PJME_MW'].mean()

# Calculating average consumption by shortened month names
monthly_avg = df.groupby('Month')['PJME_MW'].mean()

# Mapping hours
hour_labels = {
    0: "12 AM", 1: "1 AM", 2: "2 AM", 3: "3 AM", 4: "4 AM", 5: "5 AM",
    6: "6 AM", 7: "7 AM", 8: "8 AM", 9: "9 AM", 10: "10 AM", 11: "11 AM",
    12: "12 PM", 13: "1 PM", 14: "2 PM", 15: "3 PM", 16: "4 PM", 17: "5 PM",
    18: "6 PM", 19: "7 PM", 20: "8 PM", 21: "9 PM", 22: "10 PM", 23: "11 PM"
}

# Applying the mapping to rename hours
hourly_avg.index = hourly_avg.index.map(hour_labels)

# Plotting average consumption by hour
plt.figure(figsize=(8, 6))
plt.plot(hourly_avg.index, hourly_avg, marker='o', linestyle='-', color='purple', alpha=0.7)
plt.title('Average Consumption by Hour')
plt.xlabel('Hour')
plt.ylabel('Average PJME_MW')
plt.xticks(rotation=90) 
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Plotting average consumption by day of the week
dayofweek_avg = dayofweek_avg.reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.figure(figsize=(8, 6))
plt.plot(dayofweek_avg.index, dayofweek_avg, marker='o', linestyle='-', color='orange', alpha=0.7)
plt.title('Average Consumption by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average PJME_MW')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Plotting average consumption by shortened month names
month_order_short = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                     'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
monthly_avg = monthly_avg.reindex(month_order_short) 
plt.figure(figsize=(8, 6))
plt.plot(monthly_avg.index, monthly_avg, marker='o', linestyle='-', color='green', alpha=0.7)
plt.title('Average Consumption by Month')
plt.xlabel('Month')
plt.ylabel('Average PJME_MW')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Displaying the highest and lowest values in each category
print("\nHighest Average Consumption by Hour:")
print(f"{hourly_avg.idxmax()} with {hourly_avg.max():.2f} MW")
print("\nLowest Average Consumption by Hour:")
print(f"{hourly_avg.idxmin()} with {hourly_avg.min():.2f} MW")

print("\nHighest Average Consumption by Day of the Week:")
print(f"{dayofweek_avg.idxmax()} with {dayofweek_avg.max():.2f} MW")
print("\nLowest Average Consumption by Day of the Week:")
print(f"{dayofweek_avg.idxmin()} with {dayofweek_avg.min():.2f} MW")

print("\nHighest Average Consumption by Month:")
print(f"{monthly_avg.idxmax()} with {monthly_avg.max():.2f} MW")
print("\nLowest Average Consumption by Month:")
print(f"{monthly_avg.idxmin()} with {monthly_avg.min():.2f} MW")
```

> #### 1. Are there any observed trends and seasonality?
>- There is an overall decline in the average yearly load from 2002 to 2018, with peaks observed around 2006 and 2008. Strong seasonality is visible in the monthly averages, with noticeable peaks and troughs repeating every year. There are also fluctuation of patterns on daily electricity load trends. 
> #### 2. What are the peak and lowest hours, day of the week, months and explain why it might be the case?
>- <b>Hour:</b> The peak electricity consumption occurs at <b>7 PM</b>, which is likely due to increased residential and commercial activities during the evening, such as cooking, lighting, and entertainment, while the lowest consumption occurs at <b>4 AM</b>, which reflects inimal activity as most people are asleep. 
>- <b>Day of the Week:</b> <b>Tuesdays</b> experience the highest electricity consumption possibly due to regular work operations, while <b>Sundays</b> have the lowest due to reduced work and industrial activity during the weekend.
>- <b>Month:</b> <b>July</b> has the highest monthly consumption, likely driven by higher cooling demands during the summer, while <b>April</b> records the lowest consumption.

### Data Preprocessing

```python
# Extracting relevant time information
df['Month'] = df.index.month_name()
df['Day'] = df.index.day_name()
df['Hour'] = df.index.hour

# Grouping by month and calculating mean and sum
monthly_stats = df.groupby('Month')['PJME_MW'].agg(['mean', 'sum'])
monthly_stats = monthly_stats.sort_values(by='mean', ascending=False)
highest_avg_month = monthly_stats.index[0]  

# Grouping by day and calculating mean and sum
daily_stats = df.groupby('Day')['PJME_MW'].agg(['mean', 'sum'])
daily_stats = daily_stats.sort_values(by='mean', ascending=False)
highest_avg_day = daily_stats.index[0]  

# Grouping by hour and calculating mean
hourly_stats = df.groupby('Hour')['PJME_MW'].mean()
hourly_stats = hourly_stats.sort_values(ascending=False)
highest_avg_hour = hourly_stats.index[0]  

hour_labels = {
    0: "12 AM", 1: "1 AM", 2: "2 AM", 3: "3 AM", 4: "4 AM", 5: "5 AM",
    6: "6 AM", 7: "7 AM", 8: "8 AM", 9: "9 AM", 10: "10 AM", 11: "11 AM",
    12: "12 PM", 13: "1 PM", 14: "2 PM", 15: "3 PM", 16: "4 PM", 17: "5 PM",
    18: "6 PM", 19: "7 PM", 20: "8 PM", 21: "9 PM", 22: "10 PM", 23: "11 PM"
}
hourly_stats.index = hourly_stats.index.map(hour_labels)
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Hourly statistics 
axes[0].bar(hourly_stats.index, hourly_stats, color='skyblue')
axes[0].bar(hour_labels[highest_avg_hour], hourly_stats.loc[hour_labels[highest_avg_hour]], color='steelblue', label='Highest Average')
axes[0].set_title('Average Consumption by Hour')
axes[0].set_xlabel('Hour')
axes[0].set_ylabel('Average Consumption (MW)')
axes[0].legend()
axes[0].grid(axis='y', linestyle='--', alpha=0.5)  
axes[0].tick_params(axis='x', rotation=90)


# Daily statistics 
axes[1].bar(daily_stats.index, daily_stats['mean'], color='lightgreen')
axes[1].bar(highest_avg_day, daily_stats.loc[highest_avg_day, 'mean'], color='seagreen', label='Highest Average')
axes[1].set_title('Average Consumption by Day of the Week')
axes[1].set_xlabel('Day')
axes[1].set_ylabel('Average Consumption (MW)')
axes[1].legend()
axes[1].grid(axis='y', linestyle='--', alpha=0.5)  
axes[1].tick_params(axis='x', rotation=45)

# Monthly statistics 
axes[2].bar(monthly_stats.index, monthly_stats['mean'], color='yellow')
axes[2].bar(highest_avg_month, monthly_stats.loc[highest_avg_month, 'mean'], color='darkorange', label='Highest Average')
axes[2].set_title('Average Consumption by Month')
axes[2].set_xlabel('Month')
axes[2].set_ylabel('Average Consumption (MW)')
axes[2].legend()
axes[2].grid(axis='y', linestyle='--', alpha=0.5)  
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

> #### Which month/s and day of the week have the highest average consumption?
> - The highest average electricity consumption occurs at <b>7 PM</b>, on <b>Tuesdays</b>, in <b>July</b>.

### Time Series Decomposition

```python
# Resampling the data to daily frequency
df_daily = df['PJME_MW'].resample('M').mean()

# Performing decomposition
decomposition = seasonal_decompose(df_daily, model='additive')

# Extracting components
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

# Plotting the decomposition
plt.figure(figsize=(12, 10))

plt.subplot(4, 1, 1)
plt.plot(df_daily, label='Original', color='blue')
plt.title('Original Time Series')
plt.legend()

plt.subplot(4, 1, 2)
plt.plot(trend, label='Trend', color='orange')
plt.title('Trend Component')
plt.legend()

plt.subplot(4, 1, 3)
plt.plot(seasonal, label='Seasonal', color='green')
plt.title('Seasonal Component')
plt.legend()

plt.subplot(4, 1, 4)
plt.plot(residual, label='Residual', color='red')
plt.title('Residual Component')
plt.legend()

plt.tight_layout()
plt.show()
```

### ACF and PACF Plot

```python
# Resampling data to monthly
df = df.resample('M').mean(numeric_only=True)
time_series = df['PJME_MW']

fig, axes = plt.subplots(1, 2, figsize=(15, 6)) 

# Plotting ACF
plot_acf(time_series.dropna(), lags=40, ax=axes[0], title='Autocorrelation Function (ACF) - Monthly Electricity Load')
axes[0].grid()

# Plotting PACF
plot_pacf(time_series.dropna(), lags=40, ax=axes[1], title='Partial Autocorrelation Function (PACF) - Monthly Electricity Load', method='ywm')
axes[1].grid()

plt.tight_layout()
plt.show()
```

> #### Based on the plot, what model should we use (Autoregressive models (AR), Moving Averages (MA), or ARIMA?
> - The strong correlation at Lag 1 suggests that an AR(1) model could capture much of the pattern. If seasonality is suspected like periodic increases or decreases in electricity use, we can consider adding a seasonal component (like SARIMA).
